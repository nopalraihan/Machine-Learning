```python
import pandas as pd
import numpy as np

# Generate synthetic dataset
np.random.seed(0)
num_samples = 1000

# Predictor variables
age = np.random.randint(18, 65, num_samples)
income = np.random.normal(50000, 20000, num_samples)
education = np.random.choice(['High School', 'Bachelors', 'Masters', 'PhD'], num_samples)
gender = np.random.choice(['Male', 'Female'], num_samples)

# Target variable
prob_default = 1 / (1 + np.exp(-(0.1 * age + 0.0005 * income - 0.02 * (education == 'PhD') - 0.1 * (gender == 'Female') + np.random.normal(0, 0.1, num_samples))))
default = np.random.binomial(1, prob_default)

# Create DataFrame
data = pd.DataFrame({'Age': age, 'Income': income, 'Education': education, 'Gender': gender, 'Default': default})

# Map categorical variables to numeric values
data['Education'] = data['Education'].map({'High School': 0, 'Bachelors': 1, 'Masters': 2, 'PhD': 3})
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})

# Save dataset to CSV
data.to_csv('dataset.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('dataset.csv')

# Visualisasi usia terhadap default
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Default'], alpha=0.2)
plt.title('Usia terhadap Default')
plt.xlabel('Usia')
plt.ylabel('Default')
plt.grid(True)
plt.show()

# Visualisasi pendapatan terhadap default
plt.figure(figsize=(10, 6))
plt.scatter(data['Income'], data['Default'], alpha=0.2)
plt.title('Pendapatan terhadap Default')
plt.xlabel('Pendapatan')
plt.ylabel('Default')
plt.grid(True)
plt.show()

# Visualisasi pendidikan terhadap default
education_counts = data['Education'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(education_counts.index, education_counts.values)
plt.title('Pendidikan')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Jumlah')
plt.xticks(ticks=education_counts.index, labels=['High School', 'Bachelors', 'Masters', 'PhD'])
plt.grid(axis='y')
plt.show()

# Visualisasi jenis kelamin terhadap default
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=['Male', 'Female'], autopct='%1.1f%%', startangle=140)
plt.title('Jenis Kelamin')
plt.axis('equal')
plt.show()

```
